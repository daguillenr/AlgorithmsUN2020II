```
def ModifiedBFS(s): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(s) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                    node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append(d) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
def ModifiedBFSPreSolsLevel(s,L):  

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict()
        G = pydot.Dot(graph_type="digraph")
        
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append([s,0]) 
        visited[lstr(s)] = True
  
        while queue: 
  
            # Dequeue a vertex from queue 
            e = queue.pop(0)
            s = e[0]
            l = e[1] 
            graph[lstr(s)]=[]

            flag = False
            for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]
                if(d == sorted(d)):
                     flag = True
            if flag:
              node = pydot.Node(lstr(s),style="filled", fillcolor="yellow")
            else: 
              node = pydot.Node(lstr(s))

            G.add_node(node)
            
            # Get all adjacent vertices of the 
            # dequeued vertex s. If a adjacent 
            # has not been visited, then mark it 
            # visited and enqueue it 
            if(s != sorted(s)) and (l < L):
              for i in range(len(s)-1):
                d = s[i+1::-1] + s[i+2:]  
                if (d != s) and not(lstr(d) in visited.keys()):
                 
                  if(d != sorted(d)):
                     node = pydot.Node(lstr(d))
                  else:
                    node = pydot.Node(lstr(d),style="filled", fillcolor="green") # first sorted green
                  
                  G.add_node(node)
                  edge = pydot.Edge(lstr(s),lstr(d))
                  G.add_edge(edge)
                  graph[lstr(s)].append(lstr(d))
                  queue.append([d,l+1]) 
                  visited[lstr(d)] = True 
              
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

#1. Pancake Net & Pn Calcs


```
def PancakeNetwork(p):
    op = p
    n = p[-1]
    o = lstr(p)
    graph = defaultdict()
    visitado = set()
    queue = []

    queue.append(p)
    #Se aÃ±ade el nodo visitado
    visitado.add(lstr(p))
    #Ciclo while para recorrer la queue e ir graficando...
    while queue:
        p = queue.pop(0)

        graph[lstr(p)]=[]

        for i in range(len(p)-1):

            d = p[i+1::-1] + p[i+2:]

            if not(lstr(d) in visitado):
                
                queue.append(d)

                graph[lstr(p)].append(lstr(d))
                visitado.add(lstr(d))

    #Definimos la funcion de pn

    def pn(k):
        depth = -1
        for i in graph[k]:
            depth = max(depth, pn(i))
        return(depth + 1)
    maxDepth = (pn(o))
    #Definimos la funcion de profundidad maxima de nodos
    def maxDepthNodes(k, max):
        maxNodes = ""
        if(max == 0):
            return k
        for i in graph[k]:
            maxNodes += (maxDepthNodes(i, max - 1)) + " "
        return(maxNodes)
    maxNodes = maxDepthNodes(o, maxDepth).split()

    #Generar grafica
    visitado = set()
    Graphic = pydot.Dot(graph_type="graph", simplify=True)

    queue = []

    queue.append(op)
    visitado.add(lstr(op))

    while queue:
        p = queue.pop(0)

        if p == sorted(p):
            node = pydot.Node(lstr(p),style="filled", fillcolor="blue")
        else:
            node = pydot.Node(lstr(p))

        Graphic.add_node(node)

        for i in range(len(p)-1):

            d = p[i+1::-1] + p[i+2:]

            if not(lstr(d) in visitado):

                node = pydot.Node(lstr(d))
                Graphic.add_node(node)
                edge = pydot.Edge(lstr(p),lstr(d))
                Graphic.add_edge(edge)
                
                queue.append(d)

                graph[lstr(p)].append(lstr(d))
            visitado.add(lstr(p))       
    
    for i in maxNodes:
        node = pydot.Node(i,style="filled", fillcolor="red")
        Graphic.add_node(node)

    im = Image(Graphic.create_png())
    display(im)
    print("n =", n)
    print("Pn =", maxDepth)
    print("---------------")
    print("---------------")
```

```
n=5

for i in range(1, n+1):
    PancakeNetwork([*range(1, i+1)])
```
